# Making figures with numpy arrays

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
```

```python
x[:10]
```

```python
# Line plot

fig,ax=plt.subplots()
ax.plot(x,x**2);
```

```python
# Scatter 

fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# bar

nut_butter_prices={'Almond butter':10,'Peanut butter':8,"Cashew butter":12}

fig,ax=plt.subplots()
ax.set(title="Butter Prices",xlabel="Name",ylabel="Price")
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
```

```python
fig,ax=plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

```python
# Hist

x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

```python

fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax1.plot(x,x/2); 
ax2.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax3.scatter(np.random.random(10),np.random.random(10));
ax4.hist(np.random.randn(1000)); 
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10)); 
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(10000));
```

```python
heart_disease=pd.read_csv("heart-disease.csv");
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,16),subplots=True);
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind='scatter',x='age',y='chol',c='target')
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(x='age',y='chol',c='target',ax=ax,kind='scatter')
ax.axhline(y=over_50['chol'].mean(),linestyle='--')
```

```python
fig,(ax0,ax1)=plt.subplots(ncols=1,nrows=2,figsize=(10,10))

ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
ax0.set(title='Heart diseas and Cholestorol lavels',xlabel='Age',ylabel='Cholestorol')

ax0.axhline(y=over_50['chol'].mean(),linestyle='--'); 

ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])
ax1.set(title='Heart Diesease and Max Heart Rate',
       ylabel='Age',xlabel='Max Heart Rate',)
ax1.axhline(y=over_50['thalach'].mean(),linestyle='--')
```
