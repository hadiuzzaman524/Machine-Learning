```python
import pandas as pd
```

```python
# make a series for data frame

series=pd.Series(['BMW','Toyta','Honda'])
series
```

```python
# make andother series for data frame

color=pd.Series(['Black','Green','Blue'])
color
```

```python
#make a dataframe using two series

dataFrame=pd.DataFrame({'Model':series,'Colors':color})
dataFrame
```

```python
# exporting data frame

dataFrame.to_csv('newDataFrame.csv',index=False)
```

```python
# import newly created dataframe

dtfo=pd.read_csv('newDataFrame.csv')
dtfo
```
